```python
import json
from slide_agent import SlideAgent
```

```python
def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


def save_json(data, path):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)


def enrich_presentation(skeleton, user_prompt, document_text=None):
    agent = SlideAgent()

    for slide in skeleton["slides"]:
        agent.generate_slide(
            slide=slide,
            user_prompt=user_prompt,
            document_text=document_text
        )

    return skeleton
```

```python
if __name__ == "__main__":
    skeleton = load_json("input_skeleton.json")

    user_prompt = "Generate an investigation presentation about a system outage in March 2026."
    document_text = None

    enriched = enrich_presentation(skeleton, user_prompt, document_text)

    save_json(enriched, "output_generated.json")

    print("Presentation JSON generated successfully.")
```
